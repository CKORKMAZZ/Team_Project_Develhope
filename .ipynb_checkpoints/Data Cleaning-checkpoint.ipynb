## Data Cleaning

### Google DataFrame

```python
#Checking Missing values
google.isnull().sum()
```

```python
#Missing values are replaced with Not Defined
#google = google.fillna("Not Defined")
```

```python
#It can clearly be seen that there are some duplicated values in
#the dataframe that needs to be removed!
google["App"].describe()
```

```python
#Duplicated values are dropped!
google = google.drop_duplicates("App")
```

```python
google.hist(column = 'Rating',bins = 100)

plt.xlim(0,5)
plt.show()
```

```python
#It can be seen that there is a rating value which is not in the range of 0-5 
google["Rating"].plot(kind = 'box')
```

```python
#Since every value at 10472. index has shifted we observe that 19.0 as a rating value for the corresponding app
#This row needs to be removed in our Dataset!
google[(google["Rating"] < 0) | (google["Rating"] > 5)]
```

```python
google = google.drop(labels = [10472], axis = 0) 
```

```python
#Assigning mean of the rating instead of NULL values
rating_mean = np.where(google["Rating"].isnull(), np.mean(google["Rating"]), google["Rating"])
google["Rating"] = rating_mean
```

### Datatype Corrections

```python
#Data type of rating values are changed into float
google["Rating"] = google["Rating"].astype(float)
```

```python
google["Reviews"] = google["Reviews"].astype(float)
```

```python
#M values needs to be changed to corresponding numeric values
google["Size"]
```

```python
google["Size"] = google["Size"].replace("Varies with device", np.nan)
google["Size"] = google["Size"].str.replace("M","000")
google["Size"] = google["Size"].str.replace("k","")
google["Size"] = google["Size"].replace("1,000+",1000)
google["Size"] = google["Size"].astype(float)
```

```python
#'string' values in Installs columns are changed into 'float'
google["Installs"] = google["Installs"].apply(lambda x : x.strip('+').replace(',','')).astype(float)
```

```python
google["Price"].unique()
```

```python
google["Price"] = google["Price"].str.replace("$","").astype(float)
```

## Plotting

```python
google.head()
```

```python
google_sorted_cat = google.groupby("Category")["Installs"].sum().sort_values(ascending = False)
```

```python
google_sorted_rev = google.groupby("Category")["Reviews"].sum().sort_values(ascending = False)
```

```python
google_sorted_cat = google_sorted_cat.reset_index()
```

```python
google_sorted_rev = google_sorted_rev.reset_index()
```

```python
fig , ax = plt.subplots()

ax.bar('Category', 'Installs', data = google_sorted_cat.reset_index() , color = 'g')
ax.set_xlabel("Category")
ax.set_ylabel("Installations", color = 'g')
ax.tick_params('y', colors = 'g')
ax.tick_params('x' , labelrotation = 90)

ax2 = ax.twinx()

ax2.plot(google_sorted_rev["Category"],google_sorted_rev["Reviews"], color = 'r')
ax2.set_ylabel('Reviews', color = 'r')
ax2.tick_params('y', color = 'r')

plt.title("Number of Installations and Reviews By Categories")
plt.show()
```

```python
google_sorted_ins = google.sort_values(["Installs", "Rating"], ascending = [False, False]).head(10)
```

```python
google_corr = google.corr()
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(google_corr, annot = True, cmap = 'plasma')

plt.show()
```

```python
fig , ax = plt.subplots()
fig.set_size_inches = (15, 8)

popular_app_cats = google[google["Category"].isin(['GAME','FAMILY','PHOTOGRAPHY','MEDICAL',
                                                   'TOOLS', 'FINANCE', 'LIFESTYLE', 'BUSINESS'])]

ax = sns.stripplot(x = popular_app_cats["Price"], y = popular_app_cats["Category"], jitter = True, linewidth = 1)
ax.set_title('App pricing trend across categories')

plt.show()
```

```python
sns.barplot(x = "App" , y = "Rating", data = google_sorted_ins)

plt.title("Top 10 Installed and Highest Rated Products")
plt.xlabel("Applications")
plt.ylabel("Rating")
plt.tick_params(rotation = 90)
plt.show()
```

```python
google_sum_cat = google.groupby("Category")["Price"].sum().sort_values(ascending = False).reset_index()
```

```python
plt.figure(figsize = (15,10))
ax = sns.barplot(x = "Category", y = 'Price', data = google_sum_cat)


ax.set(xlabel = "Category", ylabel = "Price", title = 'Total Prices Of Each Applications By Categories')
ax.tick_params(rotation = 90)


plt.show()
```
